1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

```python
import pandas as pd
import numpy as np
```

```python
data= pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv')
#display(data)

def lower_case_spaces_replacest(df: pd.DataFrame)-> pd.DataFrame:
    cols = []
    df1=df.copy()
    for colname in df1.columns:
        cols.append(colname.lower())
    df1.columns = cols
    df1.columns = df1.columns.str.lower().str.replace('[^0-9a-zA-Z]+','_',regex=True)
    df1 = df1.rename(columns={'st':'state'})
   
    return df1

#display(lower_case_spaces_replacest(data))
data= lower_case_spaces_replacest(data)
display(data)
```

```python

    def total_claim_amount1000(df: pd.DataFrame) -> pd.DataFrame:
        filtered_df = df.query('total_claim_amount >= 1000 and response == "Yes"')
    
        return filtered_df

display(total_claim_amount1000(data))
data1= total_claim_amount1000(data)
print(data1['response'].value_counts())
print(data1['total_claim_amount'].unique())
```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
pivot1= data.pivot_table(index=['state'], values=['number_of_policies'], aggfunc='sum')
display(pivot1)
display(data['state'].value_counts())
display(data['customer'].value_counts())
#because there are clients or id's that appear several times, I cannot simply use the frequency of each state value.
#my idea is to drop those rows, but then I wouldn't be working with the original data frame anymore


customers_per_state = data.groupby('state')['customer'].nunique().reset_index()
state_500plus = customers_per_state[customers_per_state['customer'] > 500]
display(state_500plus)
```

5. The marketing team wants to analyze the number of policies sold by state and month. Present the data in a table where the months are arranged as columns and the states are arranged as rows.

```python
data['effective_to_date'] = pd.to_datetime(data['effective_to_date'])
datainsight3= data.groupby([data['effective_to_date'].dt.month, 'state'])['number_of_policies'].sum().reset_index()
datainsight31 = datainsight3.pivot_table(index='state', columns='effective_to_date', values='number_of_policies', aggfunc='sum')
display(datainsight31)
```
